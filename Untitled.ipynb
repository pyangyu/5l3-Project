```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
file_path = r'Menu.csv'
df = pd.read_csv(file_path)
```

```python
df.info(verbose=True)
```

```python
df.describe()
```

```python
for col in df.columns:
    print(f"--- Describe for {col} ---")
    print(df[col].describe())
    print("\n")
```

```python
for col in df.columns:
    print(f"--- Value counts for {col} ---")
    print(df[col].value_counts())
    print("\n")
```

```python
import pandas as pd

value_counts_dict = {}

for col in df.columns:
    counts = df[col].value_counts().to_dict()
    
    filtered_counts = {key: value for key, value in counts.items() if value > 1}
    
    value_counts_dict[col] = filtered_counts

for col, counts in value_counts_dict.items():
    print(f"Value counts for {col} (counts > 1):")
    for key, value in counts.items():
        print(f"{key}: {value}")  
    print("--" * 50)  
```

```python
import pandas as pd

filtered_df = df[(df['name'] != df['sponsor']) & (df['name'].notna()) & (df['sponsor'].notna())]

result_df = filtered_df[['id', 'name', 'sponsor']]

print(result_df)
```

```python
filtered_df = df[(df['name'] != df['sponsor']) & (df['name'].notna()) & (df['sponsor'].notna())]

result_df = filtered_df[['id', 'name', 'sponsor']]

print(result_df)
```

```python
combined_counts = df.groupby(['place', 'venue']).size()
filtered_counts = combined_counts[combined_counts > 1]
filtered_counts_df = filtered_counts.reset_index(name='count')

sorted_filtered_counts_df = filtered_counts_df.sort_values(by=['count', 'place'], ascending=[False, True])

print(sorted_filtered_counts_df)
```

```python
unique_venues_per_place = df.groupby('place')['venue'].nunique()
print(unique_venues_per_place)
```

```python
unique_venues_list = df.groupby('place')['venue'].agg(lambda x: set(x))
print(unique_venues_list)
```

```python
# Filter to show only places associated with more than one venue
multiple_venues = unique_venues_per_place[unique_venues_per_place > 1]
print(multiple_venues)

# For listing venues
multiple_venues_list = unique_venues_list[unique_venues_per_place > 1]
print(multiple_venues_list)
```
